```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path = "/content/drive/MyDrive/SPAM/UCIspamDataset/"
dataset_path = path + "phishing_dataset.csv"
data = pd.read_csv(dataset_path)
```

```python
data.shape
```

```python
sample = data.iloc[:,:-1].values
target = data.iloc[:,-1].values
X_train, X_test, y_train, y_test = train_test_split(sample, target, test_size=0.3, random_state=0)
```

```python
decision_classifier = tree.DecisionTreeClassifier()
decision_classifier.fit(X_train, y_train)
predictions = decision_classifier.predict(X_test)
matrix = confusion_matrix(y_test, predictions)
accuracy = accuracy_score(y_test, predictions) * 100
```

```python
print("Confusion matrix:\n{}".format(matrix))
print(f"Accuracy of the model is: {accuracy}")
```
